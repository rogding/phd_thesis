# Constants

```python
C = 299792458 # Speed of light in [m/s]
H = 6.62607015E-34 # Planck's constant [Js]
HBAR = H/(2*np.pi) # Planck's constant [Js]
KB = 1.380649E-23 # Boltzmann constant [J/K]
E0 = ufloat(8.8541878128, 0.0000000013)*1E-12 # Vacuum electric permittivity (epsilon_0) [F/m]
BOHR = ufloat(5.29177210903, 0.00000000080)*1E-11 # Bohr radius [m]
```

```python
wavelength = 1064E-9 # Wavelengtn in [m]
gamma = 7.5E3 # linewidth in [Hz]
f_88 = ufloat(434829121311, 10)*1000 # [Hz] from Sansonetti2010.JPCRD.39.033103
f_86 = ufloat(434828957494, 10)*1000 # [Hz] from Sansonetti2010.JPCRD.39.033103
f_84 = ufloat(434828769718, 111)*1000 # [Hz] from Hui2015.CPB.24.1.013201

#Gamma = 2*np.pi*gamma
omega_88 = 2*np.pi*f_88 # [1/s]
omega_86 = 2*np.pi*f_86 # [1/s]
omega_84 = 2*np.pi*f_84 # [1/s]
```

# General Functions

```python
def Udip(omega, omega_0, Gamma, I):
    return -(3*np.pi*C**2)/(2*omega_0**3)*(Gamma/(omega_0-omega) + Gamma/(omega_0+omega))*I

def GammaSc(omega, omega_0, Gamma, I):
    return (3*np.pi*C**2)/(2*HBAR*omega_0**3)*(omega/omega_0)**3*(Gamma/(omega_0-omega) + Gamma/(omega_0+omega))**2*I

def waist(z, z0, w0):
    zR = np.pi*w0**2/wavelength
    return np.sqrt(w0*(1+(z-z0)**2/zR**2))

def peak_intensity_circular(P, w0):
    return (2*P/np.pi/w0**2)

def intensity_circular(P, w0, z, z0, r):
    I0 = peak_intensity_circular(P, w0)
    wz = waist(z, z0, z0)
    return I0*(w0/wz)**2*np.exp(-2*r**2/wz**2)

def ac_polarizability(omega, omega_0, Gamma):
    return 6*np.pi*E0*C**3*Gamma/(omega_0**2*(omega_0**2 - omega**2)) 
```

# Simple, two-level calculations

### AC (dipole) polarizability

```python
norm = 1/(4*np.pi*E0*BOHR**3)
[delta_min, delta_max] = [-500E6, 50E6]
delta_L = np.linspace(delta_min, delta_max, num=5001, endpoint=True)
f_L = f_88 + delta_L
omega_L = 2*np.pi*f_L

plt.figure()

# 88Sr
alpha_88 = ac_polarizability(omega_L, omega_88, 2*np.pi*(7.5E3))*norm
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_88))
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_88)+unumpy.std_devs(alpha_88))
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_88)-unumpy.std_devs(alpha_88))

# 86Sr
alpha_86 = ac_polarizability(omega_L, omega_86, 2*np.pi*(7.5E3))*norm
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_86))
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_86)+unumpy.std_devs(alpha_86))
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_86)-unumpy.std_devs(alpha_86))

# 84Sr
alpha_84 = ac_polarizability(omega_L, omega_84, 2*np.pi*(7.5E3))*norm
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_84))
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_84)+unumpy.std_devs(alpha_84))
plt.plot(delta_L/1E6, unumpy.nominal_values(alpha_84)-unumpy.std_devs(alpha_84))

ax = plt.gca()
#ax.set_ylim((-2000, 2000))

ax.tick_params(direction='in', axis='both', which='both', bottom=True, top=True, left=True, right=True, length=5)



plt.tight_layout()
plt.subplots_adjust(hspace = 0)

plt.show()
```
