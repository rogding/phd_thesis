```python
data = pd.read_excel('results_digital_data.xlsx', sheet_name='transmission data')
wl = data['WL(nm)']
trans_DN16 = data['T(%) ( window 1-1/3)']
trans_DN40 = data['T(%)( window 2-3/4)']
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.gca()

color_DN40 = sns.color_palette("Paired")[1]
color_DN16 = sns.color_palette("Paired")[5]

plt.plot(wl, trans_DN16, label=r'\SI{1.33}{\inch} (DN16) viewport',
         color=color_DN16, zorder=1)

plt.plot(wl, trans_DN40, label=r'\SI{2.75}{\inch} (DN40) viewport',
         color=color_DN40, zorder=2)

# Plot annotations
fontsize=14

# 461 nm
wl_center = 461
y_center = 91.5
text_x_offset = 10
text_y_offset = -0.5

plt.axvline(x=wl_center, color='black', linewidth=1, zorder=0)
plt.text(x=wl_center + text_x_offset, y=y_center, s=r'\SI{'+str(wl_center)+r'}{\nm}',
         color='black', fontsize=fontsize)

T_DN40 = str(data.loc[data['WL(nm)']==wl_center, 'T(%)( window 2-3/4)'].item()) + r'\si{\percent}'
plt.text(x=wl_center + text_x_offset, y=y_center + text_y_offset, s=T_DN40,
         color=color_DN40, fontsize=fontsize)

T_DN16 = str(data.loc[data['WL(nm)']==wl_center, 'T(%) ( window 1-1/3)'].item()) + r'\si{\percent}'
plt.text(x=wl_center + text_x_offset, y=y_center + 2*text_y_offset, s=T_DN16,
         color=color_DN16, fontsize=fontsize)

# 689 nm
wl_center = 689
y_center = 93
text_x_offset = -10
text_y_offset = -0.5

plt.axvline(x=wl_center, color='black', linewidth=1, zorder=0)
plt.text(x=wl_center + text_x_offset, y=y_center, s=r'\SI{'+str(wl_center)+r'}{\nm}',
         color='black', fontsize=fontsize, ha='right')

T_DN40 = str(data.loc[data['WL(nm)']==wl_center, 'T(%)( window 2-3/4)'].item()) + r'\si{\percent}'
plt.text(x=wl_center + text_x_offset, y=y_center + text_y_offset, s=T_DN40,
         color=color_DN40, fontsize=fontsize, ha='right')

T_DN16 = str(data.loc[data['WL(nm)']==wl_center, 'T(%) ( window 1-1/3)'].item()) + r'\si{\percent}'
plt.text(x=wl_center + text_x_offset, y=y_center + 2*text_y_offset, s=T_DN16,
         color=color_DN16, fontsize=fontsize, ha='right')

# 698 nm
wl_center = 698
y_center = 95.5
text_x_offset = 10
text_y_offset = -0.5

plt.axvline(x=wl_center, color='black', linewidth=1, zorder=0)
plt.text(x=wl_center + text_x_offset, y=y_center, s=r'\SI{'+str(wl_center)+r'}{\nm}',
         color='black', fontsize=fontsize)

T_DN40 = str(data.loc[data['WL(nm)']==wl_center, 'T(%)( window 2-3/4)'].item()) + r'\si{\percent}'
plt.text(x=wl_center + text_x_offset, y=y_center + text_y_offset, s=T_DN40,
         color=color_DN40, fontsize=fontsize)

T_DN16 = str(data.loc[data['WL(nm)']==wl_center, 'T(%) ( window 1-1/3)'].item()) + r'\si{\percent}'
plt.text(x=wl_center + text_x_offset, y=y_center + 2*text_y_offset, s=T_DN16,
         color=color_DN16, fontsize=fontsize)

# 1064 nm
wl_center = 1064
y_center = 94
text_x_offset = 10
text_y_offset = -0.5

plt.axvline(x=wl_center, color='black', linewidth=1, zorder=0)
plt.text(x=wl_center + text_x_offset, y=y_center, s=r'\SI{'+str(wl_center)+r'}{\nm}',
         color='black', fontsize=fontsize)

T_DN40 = str(data.loc[data['WL(nm)']==wl_center, 'T(%)( window 2-3/4)'].item()) + r'\si{\percent}'
plt.text(x=wl_center + text_x_offset, y=y_center + text_y_offset, s=T_DN40,
         color=color_DN40, fontsize=fontsize)

T_DN16 = str(data.loc[data['WL(nm)']==wl_center, 'T(%) ( window 1-1/3)'].item()) + r'\si{\percent}'
plt.text(x=wl_center + text_x_offset, y=y_center + 2*text_y_offset, s=T_DN16,
         color=color_DN16, fontsize=fontsize)

# Plotting options

plt.xlim((min(wl), max(wl)))
plt.ylim((89.5, 100))

plt.xlabel(r'Wavelength [\si{\nm}]')
plt.ylabel(r'Transmission [\si{\percent}]')

# Axis options
ax.tick_params(direction='in', axis='both', which='both', bottom=True, top=True, left=True, right=True, length=6)

plt.tight_layout()
plt.legend(frameon=False, bbox_to_anchor=(0.6, 0.03), fontsize=14)

plt.savefig('viewport_ar-coating.pdf', bbox_inches='tight')

plt.show()
```
