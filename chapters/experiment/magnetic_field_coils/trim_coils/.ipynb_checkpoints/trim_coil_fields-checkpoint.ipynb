```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Constants
mu0 = (4e-7*np.pi)/10000 # [G*m/A] vacuum permeability
```

```python
# B-field due to a finite length wire symmetrical about XY, XZ, YZ planes.
# r           : vector r=(x,y,z) of the position at which to calculate field
# wire_params : 
# out         : B-field fector at specified r=[x,y,z]
def finite_x_wire(r, wire_params):
    [x, y, z] = r
    
    [L, turns, I] = wire_params
    
    I = turns*I
    
    # For wire along x-axis centered at y=0, z=0
    Bx = (mu0*I/4/np.pi)*0
    By = (mu0*I/4/np.pi)*(-2)*((L/2)*z)/((y**2 + z**2)*(np.sqrt((L/2)**2+(y**2 + z**2))))
    Bz = (mu0*I/4/np.pi)*(2)*((L/2)*y)/((y**2 + z**2)*(np.sqrt((L/2)**2+(y**2 + z**2))))
    
    
    return [Bx, By, Bz]
```

```python
x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)
z=np.linspace(-2,2,100)

X, Y, Z = np.meshgrid(x,y,z)

[Bx, By, Bz] = finite_x_wire([X,Y,Z], [1, 1, 1])
plt.streamplot(Y ,Z, By, Bz)
```
