```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
red_MOT_84Sr = pd.read_csv('red_MOT_pictures-84Sr-in_situ-with_87Sr-16471600007ODimagebatch.txt', sep='\t', header=None)
red_MOT_87Sr = pd.read_csv('red_MOT_pictures-87Sr-in_situ-no_sagging-with_84Sr-16351200010ODimagebatch.txt', sep='\t', header=None)
```

```python
fig = plt.figure(figsize=[5,5])
ax = plt.gca()
plt.pcolor(red_MOT_84Sr + red_MOT_87Sr, cmap='jet')

pixel_size = 14.1 #14.1 um/pixel
step = 25
[xmin, xmax] = [25, 175]
[ymin, ymax] = [25, 175]

ax = plt.gca()
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

xrange = np.arange(xmin, xmax + step, step)
yrange = np.arange(ymin, ymax + step, step)

[xcntr, ycntr] = [100, 100]

plt.xticks(xrange, pixel_size*(xrange - xcntr))
plt.yticks(yrange, pixel_size*(yrange - ycntr))

plt.xlabel('Horizontal [$\mu$m]')
plt.ylabel('Vertical [$\mu$m]')

ax.set_aspect('equal')

clb = plt.colorbar()
clb.ax.set_title('OD')

plt.savefig('red_MOT-84-87.svg', bbox_inches='tight')
plt.show()
```
