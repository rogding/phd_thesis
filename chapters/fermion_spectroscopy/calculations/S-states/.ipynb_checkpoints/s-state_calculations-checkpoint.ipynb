# Some general use functions

```python
# Calculate quantum defect
def qd(n, d0, d2, d4):
    return d0 + d2/(n-d0)**2 + d4/(n-d0)**4

def E_Ryd(n, I_S, RY, d):
    [d0, d2, d4] = d
    return I_S - RY/(n-qd(n, d0, d2, d4))**2
```

# Semiclassical overlap integral $<n^*|n'^*>$

Following the formula in 10.1103/PhysRevA.24.161 (cited in 10.1103/PhysRevLett.51.771).

```python
def semiclassical_overlap(n1, n1_star, n2, n2_star):
    out = (-1)**(n2-n1) * 2*np.sqrt(n1_star*n2_star)/(n1_star+n2_star) * np.sin(np.pi*(n2_star-n1_star))/(np.pi*(n2_star-n1_star))
    return out
```

```python
n = np.arange(30, 150, 1)

n1 = n
n1_star = n1-qd(n, 3.26896, -0.138, 0.9)
n2 = n
n2_star = n2-qd(n, 3.371, 0.5, -10)

y1 = semiclassical_overlap(n1, n1_star, n2, n2_star)

n1 = n
n1_star = n1-qd(n, 3.26896, -0.138, 0.9)
n2 = n+1
n2_star = n2-qd(n, 3.371, 0.5, -10)

y2 = semiclassical_overlap(n1, n1_star, n2, n2_star)

n1 = n
n1_star = n1-qd(n, 3.26896, -0.138, 0.9)
n2 = n+2
n2_star = n2-qd(n, 3.371, 0.5, -10)

y2 = semiclassical_overlap(n1, n1_star, n2, n2_star)

[fig, axs] = plt.subplots(nrows=3, ncols=1, sharex=True, figsize=(10,10))

axs[0].plot(n, np.abs(y1), '.', label=r'$\left\langle{n^*}|{n^*}\right\rangle$')

axs[1].plot(n, np.abs(y2), '.', label=r'$\left\langle{n^*}|{\left(n+1\right)^*}\right\rangle$')

axs[2].plot(n, np.abs(y2), '.', label=r'$\left\langle{n^*}|{\left(n+2\right)^*}\right\rangle$')

# Axis labels
plt.xlabel(r'$n$')
axs[0].set_ylabel(r'$\left\langle{n^*}|{n^*}\right\rangle$')
axs[1].set_ylabel(r'$\left\langle{n^*}|{\left(n+1\right)^*}\right\rangle$')
axs[2].set_ylabel(r'$\left\langle{n^*}|{\left(n+2\right)^*}\right\rangle$')

#plt.legend()

# Tick settings
for ax in axs:
    ax.tick_params(axis='both', which='both', direction='in')
    ax.xaxis.set_ticks_position('both')
    ax.yaxis.set_ticks_position('both')

# Remove space between subplots
plt.subplots_adjust(wspace=0, hspace=0.05)

plt.show()
```
