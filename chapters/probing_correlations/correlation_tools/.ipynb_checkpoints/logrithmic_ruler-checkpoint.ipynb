```python
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# https://stackoverflow.com/questions/22348756/upright-mu-in-plot-label-retaining-original-tick-fonts
# https://stackoverflow.com/questions/2537868/sans-serif-math-with-latex-in-matplotlib

import matplotlib
from matplotlib import rc

matplotlib.rc('text', usetex = True)
matplotlib.rc('font', **{'family':'sans-serif'})

params = {'text.latex.preamble':[r'\usepackage{siunitx}',
                                 r'\usepackage{amsmath}',
                                 r'\usepackage{physics}',
                                 r'\usepackage{helvet}',
                                 r'\usepackage{sansmath}',
                                 r'\sansmath',
                                 r'\sisetup{detect-all}']}
plt.rcParams.update(params)
```

```python
sns.set_context('talk')

#with sns.plotting_context('talk'), sns.axes_style('ticks'):
a0 = (5.29177210903E-11)*(1E9) # [nm] Bohr radius

#savekwargs = {'bbox_inches':'tight','dpi':500}

fig, ax = plt.subplots(figsize=(10,7))
ax2 = ax.twiny()

ax.set_xlabel(r'Interparticle separation ($R$) [\si{\nm}]')
ax2.set_xlabel(r'Interparticle separation ($R$) [\si{\bohr}]')

ax.set_xscale('log')
ax2.set_xscale('log')

ax.set_xlim(0.5,2500)
ax2.set_xlim(ax.get_xlim()[0]/a0,ax.get_xlim()[1]/a0)

ax2.tick_params(which='both',direction='in',pad=0)
ax.tick_params(which='both',direction='in',pad=3)
ax.set_ylim(0,1)
ax.yaxis.set_ticks([])
#f.savefig('blank.png',**savekwargs)

###############################
# Constants
text_up = 0.04 # Text above
text_dw = -0.04 # Text above

###############################
# Length scales

# Densities
[rho_min, rho_max] = [1E12, 1E15]
r_min = ((5E6)*(6/np.pi)**(1/3))/rho_min**(1/3) # [nm] 10^12 /cm^3
r_max = ((5E6)*(6/np.pi)**(1/3))/rho_max**(1/3) # [nm] 10^15 /cm^3
x_pos = np.sqrt(r_min*r_max)
y_pos = 0.85
ax.hlines(y=y_pos, xmin=r_min, xmax=r_max, color='black')
ax.annotate(xy=(x_pos, y_pos + text_up),s=r'$\pqty*{3/4 \pi \rho}^{1/3}$', ha='center', va='center')
ax.annotate(xy=(r_min, y_pos + text_dw), s=r'\SI{E12}{\per\cm\cubed}', ha='center', va='center')
ax.annotate(xy=(r_max, y_pos + text_dw), s=r'\SI{E15}{\per\cm\cubed}', ha='center', va='center')

# Thermal de Broglie wavelength
r_min = 186.105 # [nm] for 1 uK 88Sr atoms
r_max = 588.516 # [nm] for 100 nK 88Sr atoms
x_pos = np.sqrt(r_min*r_max)
y_pos = 0.7
ax.hlines(y=y_pos, xmin=r_min, xmax=r_max, color='black')
ax.annotate(xy=(x_pos, y_pos + text_up),s=r'$\lambda_{\text{dB}}$', ha='center', va='center')
ax.annotate(xy=(r_min, y_pos + text_dw), s=r'\SI{1}{\micro\kelvin}', ha='center', va='center')
ax.annotate(xy=(r_max, y_pos + text_dw), s=r'\SI{100}{\nano\kelvin}', ha='center', va='center')

# Optical lattices
r_min = 532/2 # [nm]
r_max = 1064/2 # [nm]
x_pos = np.sqrt(r_min*r_max)
y_pos = 0.5
ax.hlines(y=y_pos, xmin=r_min, xmax=r_max, color='black')
ax.annotate(xy=(x_pos, y_pos + text_up),s=r'Optical lattices', ha='center', va='center')

# van der Waals length scales
r_min = 2 # [nm] from TK's plot
r_max = 5 # [nm] from TK's plot
x_pos = np.sqrt(r_min*r_max)
y_pos = 0.4
ax.hlines(y=y_pos, xmin=r_min, xmax=r_max, color='black')
ax.annotate(xy=(x_pos, y_pos + text_up),s=r'$R_{\text{vdW}}$', ha='center', va='center')

# Sr resonant scattering
r_min = 798*a0 # [nm] 86Sr-86Sr at 798 a0
r_max = 1658*a0 # [nm] 88Sr-84Sr at 1658 a0
x_pos = np.sqrt(r_min*r_max)
y_pos = 0.4
ax.hlines(y=y_pos, xmin=r_min, xmax=r_max, color='black')
ax.annotate(xy=(x_pos, y_pos + text_up),s=r'Sr resonant scattering', ha='center', va='center')

###############################
# Tools for probing correlations

# 3-body loss
y_pos = 0.1
ax.annotate(s=r'3-body loss ($R \simeq 0$)', xy=(.5, y_pos), xytext=(5, y_pos),
            ha='center', va='center', arrowprops={'arrowstyle':'simple', 'color':'red', 'linewidth':0.0005})

# Photoassociative spectroscopy (PAS)
r_min = 5 # [nm] from TK's plot
r_max = 10 # [nm] from TK's plot
x_pos = np.sqrt(r_min*r_max)
y_pos = 0.25
ax.hlines(y=y_pos, xmin=r_min, xmax=r_max, color='red')
ax.annotate(xy=(x_pos, y_pos + text_up),s=r'PAS', ha='center', va='center')

#f.savefig('short_range.png', **savekwargs)

# Quantum gas microscopes (QGM)
r_min = 532/2 # [nm]
r_max = 1064/2 # [nm]
x_pos = np.sqrt(r_min*r_max)
y_pos = 0.3
ax.hlines(y=y_pos, xmin=r_min, xmax=r_max, color='red')
ax.annotate(s=r'QGM',xy=(x_pos, y_pos + text_up), ha='center', va='center')
#f.savefig('long_range.png',**savekwargs)

# Ultra-long-range Rydberg molecules (ULRRMs)
y_pos = 0.2
ax.hlines(y=y_pos, xmin=1.8*(20-3.37)**2*a0, xmax=1.8*(100-3.37)**2*a0, color='blue')
ax.annotate(s=r'ULRRM', xy=(150, y_pos + text_up), ha='center', va='center')
ax.annotate(s=r'$n=20$', xy=(1.8*(20-3.371)**2*a0, y_pos + text_dw), ha='center', va='center')
ax.annotate(s=r'$n=100$', xy=(1.8*(100-3.371)**2*a0, y_pos + text_dw), ha='center', va='center')

###############################
# Saving figure
fig.savefig('correlation_length_scales.pdf', bbox_inches='tight')
fig.savefig('correlation_length_scales.png', bbox_inches='tight', dpi=1000)

plt.show()
```
