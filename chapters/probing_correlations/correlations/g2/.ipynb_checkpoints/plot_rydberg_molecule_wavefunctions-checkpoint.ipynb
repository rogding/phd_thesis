```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
potential = np.genfromtxt('pot_87Sr_with_ion_atom_n34.txt', skip_header=2)
nu0 = np.genfromtxt('wvf_87Sr_with_ion_atom_n34-nu_0.txt',skip_header=2)
nu1 = np.genfromtxt('wvf_87Sr_with_ion_atom_n34-nu_1.txt',skip_header=1)
nu2 = np.genfromtxt('wvf_87Sr_with_ion_atom_n34-nu_2.txt',skip_header=1)
nu3 = np.genfromtxt('wvf_87Sr_with_ion_atom_n34-nu_3.txt',skip_header=1)
nu4 = np.genfromtxt('wvf_87Sr_with_ion_atom_n34-nu_4.txt',skip_header=1)

potential = pd.DataFrame(potential, columns=['R','E'])
nu0 = pd.DataFrame(nu0, columns=['R', 'RPsi', 'E'])
nu1 = pd.DataFrame(nu1, columns=['R', 'RPsi', 'E'])
nu2 = pd.DataFrame(nu2, columns=['R', 'RPsi', 'E'])
nu3 = pd.DataFrame(nu3, columns=['R', 'RPsi', 'E'])
nu4 = pd.DataFrame(nu4, columns=['R', 'RPsi', 'E'])
```

```python
nu_list = np.array([[nu0, r'$\nu=0$'],
                    [nu1, r'$\nu=1$'],
                    [nu2, r'$\nu=2$'],
                    [nu3, r'$\nu=3$'],
                    [nu4, r'$\nu=4$']])

sns.set_context('talk')

with sns.color_palette('bright'):
    plt.figure(figsize=(9,6))
    plt.plot(potential['R'], potential['E'], 'k', label='_nolegend_')

    for nu, l in nu_list:
        plt.plot(nu['R'], nu['RPsi']*1e2 + nu['E'], label=l)

    plt.xlim(10, max(potential['R']))
    plt.ylim(-32, 15)
    plt.ylabel('POTENTIAL ENERGY [MHz]')
    plt.gca().xaxis.set_label_position('top') 
    
    plt.xlabel('INTERNUCLEAR DISTANCE [nm]')
    
    plt.legend(frameon=False, ncol=4, loc=8)
    
    plt.gca().xaxis.tick_top()
    plt.gca().xaxis.set_ticks_position('both')
    plt.gca().yaxis.set_ticks_position('both')
    plt.gca().tick_params(direction='in')
    
    #plt.savefig('n34_wavefunctions.pdf',bbox_inches='tight', transparent=True)
    
    plt.show()
```

```python
sns.palplot(sns.color_palette('Paired'))
```

```python
nu_list = np.array([[nu0, r'$\nu=0$', sns.color_palette("Paired")[1], sns.color_palette("Paired")[0]+(0.5,)],
                    [nu1, r'$\nu=1$', sns.color_palette("Paired")[3], sns.color_palette("Paired")[2]+(0.5,)],
                    [nu2, r'$\nu=2$', sns.color_palette("Paired")[5], sns.color_palette("Paired")[4]+(0.5,)],
                    [nu3, r'$\nu=3$', sns.color_palette("Paired")[7], sns.color_palette("Paired")[6]+(0.5,)],
                    [nu4, r'$\nu=4$', sns.color_palette("Paired")[9], sns.color_palette("Paired")[8]+(0.5,)]])

wvf_scale = 9E1

sns.set_context('talk')

plt.figure(figsize=(10,6))

plt.plot(potential['R'], potential['E'], 'k', label='_nolegend_')

for nu, l, mc, fc in nu_list:
    plt.plot(nu['R'], nu['RPsi']*wvf_scale + nu['E'], label=l, color=mc)
    plt.fill_between(nu['R'], nu['E'], nu['RPsi']*wvf_scale + nu['E'], color=fc)

plt.xlim(15, max(potential['R']))
plt.ylim(-31, 15)
plt.ylabel(r'Energy [MHz]')
plt.gca().xaxis.set_label_position('bottom') 

plt.xlabel(r'Internuclear distance ($R$) [nm]')

#plt.legend(frameon=False, ncol=4, loc=8)
#plt.legend(framealpha=1, loc='lower left')
plt.legend(fontsize=16, ncol=5, handlelength=1, loc='lower center')

#plt.gca().xaxis.tick_top()
plt.gca().xaxis.set_ticks_position('both')
plt.gca().yaxis.set_ticks_position('both')
plt.gca().tick_params(direction='in')

plt.legend(frameon=False, ncol=3)

#plt.savefig('n34_wavefunctions.pdf', bbox_inches='tight', transparent=False)
plt.savefig('n34_wavefunctions.png', bbox_inches='tight', dpi=1000, transparent=False)

plt.show()
```

## Making the g^(2) inset figure

```python
sns.palplot(sns.color_palette('bright'))
```

```python
def g2(epsilon, x):
    return 1 + epsilon*np.exp(-2*np.pi*x**2)

sns.set_context('talk')
plt.figure(figsize=(6,4))

x = np.linspace(0, 1, num=1000, endpoint=True)
y_boson = g2(1, x)
y_classical = g2(0, x)
y_fermion = g2(-1, x)

plt.plot(x, y_boson, label=r'Bosons ($\epsilon=+1$)', color='blue')
plt.plot(x, y_classical, label=r'Classical ($\epsilon=0$)', color='green')
plt.plot(x, y_fermion, label=r'Fermions ($\epsilon=-1$)', color='red')

plt.xlabel(r'INTERPARTICLE SEPARATION [$R/\lambda_{\rm{dB}}$]')
plt.ylabel(r'$g^{\left(2\right)}\left(R\right)$')

plt.legend(handlelength=1, fontsize=14)

plt.gca().xaxis.set_ticks_position('both')
plt.gca().yaxis.set_ticks_position('both')
plt.gca().tick_params(direction='in')

plt.xlim(min(x), max(x))
plt.ylim(-0.1, 2.1)

plt.xticks([0, 0.25, 0.5, 0.75, 1])
plt.yticks([0, 1, 2])

plt.savefig('g2.pdf', bbox_inches='tight', transparent=True)

plt.show()
```
