```python
import os.path
import numpy as np
import pandas as pd
from uncertainties import unumpy
```

```python
# Constants
SoL = 299792458

# Names of data files in source/ subdirectory
DATA_FILES = [os.path.join('source','cou2018-table_2.csv'),
              os.path.join('source','cou2018-table_3.csv'),]

# Final DataFrame output column names
COLUMNS = {'Series':       str(),
           'n':            int(),
           'Term':         str(), 
           'Label':        str(),
           'E_exp':        float(),
           'E_exp_unc':    float(),
           'Isotope':      int(),
           'Ref':          str()}
```

```python
df_output = pd.DataFrame(columns=COLUMNS)

for file in DATA_FILES:
    temp = pd.read_csv(file, header = 0)
    
    energy = unumpy.uarray(temp['E_exp'], temp['E_exp_unc'])
    energy *= (1/SoL)*(1/100)*(1e6)
    
    if 'S' in temp['Series']:
        print('5sns')
```
