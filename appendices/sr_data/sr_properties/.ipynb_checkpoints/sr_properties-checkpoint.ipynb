# Constants and measured values

```python
C = 299792458 # Speed of light [m Hz]

# Constants from CODATA 2014

m_e = ufloat(9.10938356, 0.00000011)*1E-31 # Electron mass [kg]
m_u = ufloat(1.660539040, 0.000000020)*1E-27 # Atomic mass constant [kg]
m_p = ufloat(1.672621898, 0.000000021)*1E-27 # Proton mass [kg]

R_infty = ufloat(10973731.568508, 0.000065)/100 # Rydberg constant [/cm]

# Atomic masses from NIST
m_88 = ufloat(87.9056125, 0.0000012)*m_u
m_87 = ufloat(86.9088775, 0.0000012)*m_u
m_86 = ufloat(85.9092606, 0.0000012)*m_u
m_84 = ufloat(83.9134191, 0.0000013)*m_u
```

# Mass-scaled Rydberg constants

```python
def calc_scaled_Rydberg_constant(mass):
    R_out = R_infty*mass/(m_e + mass)
    print(r'R = ' + str(R_out))
    return R_out

R_88 = calc_scaled_Rydberg_constant(m_88)
R_87 = calc_scaled_Rydberg_constant(m_87)
R_86 = calc_scaled_Rydberg_constant(m_86)
R_84 = calc_scaled_Rydberg_constant(m_84)
```
