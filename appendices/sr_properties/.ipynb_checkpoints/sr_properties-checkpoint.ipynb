```python
from scipy import constants as spc 
from uncertainties import ufloat, ufloat_fromstr
```

```python
# CODATA AMU in [kg]
AMU = ufloat_fromstr('1.660539040(20)')*(10**-27)

# From NIST's Physical Measurement Laboratory's Elemental Data Index
# Relative atomic mass of 12C in [u]
mr12C = ufloat_fromstr('12.0000000(00)')
m12C = mr12C*AMU

# Relative atomic masses (relative to 12C = 12)
mr88Sr = ufloat_fromstr('87.9056125(12)')
mr87Sr = ufloat_fromstr('86.9088775(12)')
mr86Sr = ufloat_fromstr('85.9092606(12)')
mr84Sr = ufloat_fromstr('83.9134191(13)')

m88 = mr88Sr*(m12C/12)
m87 = mr87Sr*(m12C/12)
m86 = mr86Sr*(m12C/12)
m84 = mr84Sr*(m12C/12)

print(m88/AMU)
print(m87/AMU)
print(m86/AMU)
print(m84/AMU)
```
